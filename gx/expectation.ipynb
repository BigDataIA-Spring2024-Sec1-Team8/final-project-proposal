```python
import pandas as pd
from great_expectations.dataset import PandasDataset
```

```python
df = pd.read_csv('../products.csv')
df = df.drop_duplicates('asin')
```

```python

ge_df = PandasDataset(df)

# Define expectations
ge_df.expect_column_values_to_be_of_type('description', 'str')  # Expect 'product' column to be text
ge_df.expect_column_values_to_be_of_type('title', 'str')  # Expect 'product' column to be text
ge_df.expect_column_values_to_be_unique('asin')  # Expect 'asin' column to be unique
ge_df.expect_column_values_to_not_be_null('imageurl')  # Expect 'image' column to not have null values

cats_desired = ['Men Clothing', 'Women Clothing', 'Boys Clothing','Girls Clothing',
'Men Shoes', 'Women Shoes', 'Boys Shoes','Girls Shoes',
'Men Watches', 'Women Watches', 'Boys Watches','Girls Watches']

ge_df.expect_column_values_to_be_in_set('category', cats_desired)

# Run validation
results = ge_df.validate()

# Print validation results
print(results)
```
